# Gradient Descent

The gradient descent algorithm is the fundamental algorithm of optimization, and the foundation for modern machine and deep learning. It's a classic application of multivariable calculus, and we use it everywhere. The basic idea is pretty simple, just "go downhill".

### Review of Derivatives



To take you back to your calculus days, or to get you caught up, a derivative is an operator that tells you the slope of a function at a given point. If you put a flat line up against your function, is it pointing up or down? And how steep is it? It represents rise over run, or how much y changes with a change in x.

You can do the same thing in multiple dimensions too. If you have a function that takes in x and y and spits out z, you can still take derivatives to see where the function is pointing up or down. If you put a flat plane up against your function in three dimensions, at what angle would the plane sit? And what direction points the steepest uphill? The gradient of a function tells you just that, the direction you would need to travel to go uphill the steepest. So if you go in the opposite direction, or the negative gradient, you'll be going downhill as fast as you can.

### Summary of Optimization and the Gradient Descent Algorithm

Optimization is how we take a function with lots of inputs and one output, and see what combination of inputs would give us the biggest or the smallest output. For example, maybe you operate a toy company and different toys cost different amounts and also make different amounts of money when you sell them. You could ask, how many of each toy should I make to maximize profit? Or maybe you're grocery shopping on a budget and ask, how much of which foods will help me meet my nutritional goals for the cheapest?

In general, you set up an optimization problem by writing a function of inputs that gives you one output, called the objective function, and try to minimize the output. If you are trying to maximize something, like profit, you can always negate your function and convert it to a minimization problem. The gradient descent algorithm helps find the minimum of the objective function. You start with a guess for your starting point and take the gradient of the function to find where it points the steepest up and where it points the steepest down. Then you take a small step in the downward direction, and do it again. At each iteration you'll find the direction to walk that goes the steepest downhill and take a step in that direction. After you take enough steps you'll reach the bottom, where the gradient is flat and you'll settle into one spot, the minimum.



There are lots of nuances to making sure you find the minimum in a reasonable amount of time. If you're pretty far away from the minimum, you want to take big steps or it will take a long time to get there, but if you're close you want to take small steps so you don't overshoot and have to come back. The size of the step, sometimes called the learning rate, often starts bigger and gets smaller as you go, hoping you're getting closer and closer to the minimum. Sometimes there are multiple minima, and one might be lower than another. Maybe your function has two bowl shapes in it and one of them is higher. Gradient descent will point you downhill into the first bowl but by definition it only goes down, so it's impossible to get out of the bowl and into the second bowl. People solve this by picking lots of starting points to see where they converge, or by adding "momentum" so that you can "roll" out of small bowls and into the big ones.

### Application: Challenger Space Shuttle

In 1986, the Challenger space shuttle exploded just minutes after launch when the rubber O-rings that seal joints in the engine were damaged, leaking fuel. One hypothesis is that if it's too cold outside during a rocket launch, it could damage the O-rings and cause an explosion. We can see if that's true by solving a logistic regression problem. Logistic regression is a function that models the probability of an event happening using the sigmoid function: $S(x_1,...,x_n)=\frac{1}{1+\exp(-(\beta_0 + \beta_1 x_1 + ... + \beta_n x_n))}$. Your inputs are the $x_1$ through $x_n$ and the output is the probability of the event occurring. The $\beta$ parameters tune the function to the true probability. In this case, our objective function is to find the $\beta$s that will make our sigmoid probability as close as possible to the true probability.

If we take data from lots of rocket launches, where our input is the temperature on the day of the launch and the output is whether or not there was damage to the O-rings, our sigmoid function looks like $S(x)=\frac{1}{1+\exp(-(\beta_0 + \beta_1 x))}$ and our data has $x$, the temperature, and $y$, which is 1 if there was O-ring damage and 0 if there wasn't. We have a list of data points, temperature and whether or not there was damage, and we want to find the $\beta$s that get the outputs of our sigmoid function as close as possible to the outputs in real life. So our objective function is to minimize $|S(x) - y|$.

The result is a function that will tell you the probability of O-ring damage given the temperature that day. In the graph below we have the datapoints that we observed, and we see where the 1986 Challenger launch sits with a temperature of just 31 degrees.

```python
from modules.gradient_descent_challenger import LogisticRegression1D
from autograd import numpy as np
import matplotlib.pyplot as plt

#load x,y, fit
x,y = np.load('data/challenger.npy').T
LR = LogisticRegression1D()
LR.fit(x,y,np.array([20., -1.]))

#plot
plt.title('Probability of O-Ring Damage')
plt.xlabel('Temperature')
plt.ylabel('O-Ring Damage')
plt.plot(x,y,'o',label='Previous Damage')
domain = np.linspace(30,100,100)
plt.plot(domain, [LR.predict(x) for x in domain])
plt.plot(31,LR.predict(31),'o', label='P(Damage) at Launch')
plt.legend()
plt.show()

print(f'Probability of O-ring damage during 1986 launch: {LR.predict(31):.4f}')
```
