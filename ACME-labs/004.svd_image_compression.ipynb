# Singular Value Decomposition and Image Compression

Singular Value Decomposition, or SVD, is one of the most fundamental applications of linear algebra and is often used in dimensionality reduction and data compression. If you have data with too many features or dimensions, it will be hard to do math on your data in an efficient way, even on the fastest computer. Luckily, not all of those dimensions are vital to the content of your data. Say you have a survey with 1000 questions on it, including your age, height, weight, resting heart rate, and caloric intake. It might not be necessary to collect data on all of these questions because they're all fairly related characteristics. It might be true that if I know how old you are and how tall you are, I could probably make a good guess at your weight and caloric intake. And as long as that guess is good enough, I could just delete those features of my data instead of trying to use all 1000 questions in my data analysis. The question is, which ones can I delete?



### Coordinate Systems

There's more than one way to describe the location of a point on a plane. We usually use Euclidean coordinates, or $x$ and $y$ position. A vector $(2,4)$ means go two units to the right and four units up. But we don't have to use the horizontal and vertical axes, we could use diagonal axes instead, or any two lines through the origin for that matter as long as they don't point in the same direction. Euclidean coordinates use the standard basis vectors, $e_1=(1,0)$ and $e_2=(0,1)$. Take $2$ of $e_1$ and $4$ of $e_2$ and add them together to get $(2,4)$. But if you were using diagonal lines as your basis, $b_1=(1,1)$ and $b_2=(-1,1)$, then you would need to add $3$ of $b_1$ and $1$ of $b_2$ to get $(2,4)$. In this new basis, your point isn't $(2,4)$, it's $(3,1)$.

So say you have a bunch of 2 dimensional data that's mostly clustered around one of those diagonal lines. If you used the standard basis vectors, you would need 2 numbers for each of those data points to describe where they are on the plane. But if you picked a basis that went right through all your data, you could probably get away with just one coordinate because it would still describe where the point is. Instead of measuring $x$ and $y$, just measure how far up or down the diagonal you need to go. Now you can represent two dimensional data with just one dimension.

### Choosing a Basis

Choosing a basis to represent your data can make a huge difference in reducing the dimension or size of your data. The goal is usually to find the dimensions where your data is the most spread out, that way those dimensions can help you tell your data points apart. We do that using eigenvectors.

An eigenvector is a vector $x$ that, when multiplied by a matrix, just results in a scaled version of itself, $Ax=\lambda x$. You can think of a matrix as a transformation that rotates, stretches, and shrinks vectors in different directions. Here's what it looks like when you take a bunch of points that make up a smile and multiply them by the matrix 
$A=\begin{bmatrix}
0.8 & 1 \\
-0.8 & 1
\end{bmatrix}$.

```python
import numpy as np
import matplotlib.pyplot as plt

# load our data
A = np.array([[.8,1],[-.8,1]])
smile = np.load('data/smile.npy')

# plot just the smile
plt.figure(figsize=(6,3))
plt.subplot(121)
plt.title('Original Smiley Face')
plt.scatter(*smile)    # just plotting the points
plt.axis('equal')

# plot the transformed smile
plt.subplot(122)
plt.title('Rotated and Stretched Smiley Face')
plt.scatter(*A@smile)  # plotting A times the points
plt.axis('equal')
plt.tight_layout()
plt.show()
```

Usually multiplying a matrix by a vector rotates and stretches the vector, but if multiplying a matrix by a vector results in just a scaled version of that vector, that means it's not being roated at all. The eigenvectors represent the directions in which the matrix scales the data, and the scaling factor $\lambda$, called the eigenvalue, represents how much the matrix scales the data in that direction. So if we're trying to find the dimensions where the data is the most spread out, we can use the eigenvectors with the biggest eigenvalues.

### Singular Value Decomposition

Singular Value Decomposition finds the eigenvectors and corresponding eigenvalues that represent where our data is most spread out. First you take your matrix $A$ and multiply it by it's transpose $A^T$ (transposing means swapping the rows for the columns), which ensures that it's a square matrix and that your eigenvalues will always be positive numbers. Then you put the eigenvectors of $AA^T$ into the columns of a matrix $U$ and the eigenvectors of $A^TA$ into the columns of a matrix $V$. The singular values are the square roots of the eigenvalues, and they go along the diagonal of a matrix $\Sigma$. When you put them together, you can reconstruct $A$ using those eigenvector matrices, $A=U\Sigma V^T$. Usually, you sort the singular values in your diagonal matrix from biggest to smallest. Now you have a way to represent your data in terms of the dimensions they're stretched out, and you've sorted the dimensions by which ones are the most stretched out. Let's say you only want the biggest 5 dimensions corresponding with the biggest 5 singular values. Then take the first 5 columns of $U$, the first 5 singular values, and the first 5 columns of $V$, and when you multiply them together you get a matrix that looks a lot like your original matrix, but you're only using 5 dimensions.

Going back to the 1000 question survey, you can put people's responses to the survey into columns, where each column is a person and each row is that person's answer to a question. If you have 50 people take the survey, you have a matrix with 1000 rows and 50 columns. Instead of keeping track of all 1000 questions, we can use the SVD to find 5 different combinations of answers to questions that will best represent their actual answers to all 1000 questions. Those eigenvectors might say to take 3 times their answer to the first question, -1 times their answer to the second question and .5 times their answer to the third question, and so on. When you add all those up, you get one number that represents a combination of answers from all their questions in an "eigen-dimension". And we made sure that this "eigen-dimension" is one where the data is spread out, so knowing their value in this "eigen-dimension" actually tells us a lot about their answers to all the questions.

### Image Compression
