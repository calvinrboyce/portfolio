# Networks and the Kevin Bacon Number

Graph Theory is a field of mathematics that studies networks and how they behave. A graph in math is a collection of nodes and edges, which connect the nodes to each other. Nodes are often drawn as circles and edges are drawn as lines connecting the nodes. The edges can be directed and only point in one direction, or undirected and point in both directions, and they can have weights assigned to them. Imagine a map of the cities in America. Some of the cities have roads connecting them while others don't. You could represent this network as a graph, where each city is a node and each road is an edge. Most of the roads between cities go both directions, so you might use an undirected graph, meaning all the edges go both ways. The weight you assign to each edge could represent how long the road is from one city to the next.



There are lots of questions you could ask about graphs, but one of the fundamental questions in graph theory is finding the shortest path between two nodes. If you want to get from Boston to LA, which way should you go? There are lots of combinations of paths: you could go through Chicago, Denver, San Francisco and down, or Chicago, Denver, Dallas and over, or Atlanta, Dallas and over, etc. There are two primary algorithms to finding all the paths between two nodes on a graph, Depth First Search and Breadth First Search. We'll focus on Breadth First Search because it does better at finding the shortest path.

You'll need a list $V$ of cities you've already visited, a queue $Q$ of nodes to visit soon in the order you want to visit them, and a set $M$ of marked cities that have already been visited or that you've already planned to visit. As a reminder, a list is a bunch of things in order and you can add or take off things from either side. A queue is a list but you can only add to the back and take off the front. A set is a group of things that don't have an order. Our algorithm consists of three steps:

1. Add all neighboring nodes to our queue $Q$ and our marked set $M$ from left to right
2. Take the first node in $Q$ and add it to our visited list $V$
3. Repeat by adding neighboring nodes to $Q$ and $M$ as long as they're not already in $M$

Basically you look at all the neighboring cities and check if you've visited them yet or if you've slotted them to be visited soon. If not, add them to the queue of cities to visit, in order, then go to the next city you've marked for visiting. So in our case it would look like this:

1. Start in Boston, mark that we've been here, add NYC to our queue and marked set:  
$V:[Boston]$  
$Q:[NYC]$  
$M:\{Boston, NYC\}$  

2. Go to NYC and add DC to our queue and marked set:  
$V:[Boston, NYC]$  
$Q:[DC]$  
$M:\{Boston, NYC, DC\}$  

3. Go to DC, add Chicago and Atlanta to our queue and marked set:  
$V:[Boston, NYC, DC]$  
$Q:[Chicago, Atlanta]$  
$M:\{Boston, NYC, DC, Chicago, Atlanta\}$  

4. Go to Chicago since it's the first in our queue, add Seattle and Denver to our queue, but don't add Atlanta, even though there is a road from Chicago to Atlanta, because it's already in our marked set to visit soon:  
$V:[Boston, NYC, DC, Chicago]$  
$Q:[Atlanta, Seattle, Denver]$  
$M:\{Boston, NYC, DC, Chicago, Atlanta, Seattle, Denver\}$  

5. Go to Atlanta, add Dallas and Miami to our queue but not Chicago or DC:  
$V:[Boston, NYC, DC, Chicago, Atlanta]$  
$Q:[Seattle, Denver, Dallas, Miami]$  
$M:\{Boston, NYC, DC, Chicago, Atlanta, Seattle, Denver, Dallas, Miami\}$  

And keep repeating the pattern until you find all the paths from Boston to LA. You'll notice that after we visit Chicago we don't visit its neighbors in Seattle and Denver until we go back and visit Atlanta. That's the nature of the Breadth First Search, we'll search wide before we search deep. Once we've found all the paths from Boston to LA, we can see which one is the shortest and take that one.

Note: there are more efficient ways to find the shortest path than to find all the paths and just take the smallest one, especially if there are a lot of possible paths. For a faster way to find the shortest path, see Dijkstra's Algorithm, which allows you to eliminate extra paths you already know are going to be too long.

### Kevin Bacon Number
The "Kevin Bacon Number" comes from Hollywood, it basically measures how close actors are to being in a movie with Kevin Bacon. If you were actually in a movie with Kevin Bacon, your Kevin Bacon number is 1, but if you were in a movie with Jennifer Lawrence, and she was in a movie with Kevin Bacon, you'd have a Kevin Bacon number of 2. This is a shortest path problem and can be solved with graph theory. Here we read in actor lists from movies on IMDB to make a graph showing who played a role in what movie:

```python
import networkx as nx
import matplotlib.pyplot as plt
from modules.graphs_bfs_bacon_number import MovieGraph

# draw three movies from the graph
graph = MovieGraph(n_movies=3)
nx.draw(graph.g)
plt.show()
```

In the graph above we have three movies and the actors in those movies. In one cluster you can see there's a movie that doesn't share any actors with the other two. In the other, there are plenty of actors that only show up in one of the movies, but you can see there are several actors that show up in both movies.  

With this graph, we can use our BFS algorithm to find the shortest path between any two actors to get our Kevin Bacon number:

```python
graph = MovieGraph()

# Jennifer Lawrence
KB_number, path = graph.path_to_actor('Jennifer Lawrence', 'Kevin Bacon')
print(f"Jennifer Lawrence's Kevin Bacon number: {KB_number}\nPath: {path}\n")

# Johnny Depp
KB_number, path = graph.path_to_actor('Johnny Depp', 'Kevin Bacon')
print(f"Johnny Depp's Kevin Bacon number: {KB_number}\nPath: {path}\n")

# Donia Hames
KB_number, path = graph.path_to_actor('Donia Hames', 'Kevin Bacon')
print(f"Donia Hames's Kevin Bacon number: {KB_number}\nPath: {path}\n")

# Gordon Walsh
KB_number, path = graph.path_to_actor('Gordon Walsh', 'Kevin Bacon')
print(f"Gordon Walsh's Kevin Bacon number: {KB_number}\nPath: {path}\n")
```

As you can see, most actors have a pretty small Kevin Bacon Number, and the Six Degrees of Kevin Bacon game is to try and find an actor with a Kevin Bacon number higher than 6. With our new graph, we can actually plot a histogram of all the Kevin Bacon numbers and see that there are no actors with a Kevin Bacon number above 6.
